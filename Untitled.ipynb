# LINE PLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd



var =[1,2,3,4,5,6,7]
var_1 =[2,3,4,1,5,6,7]
```

```python
plt.plot(var,var_1)
plt.show()
```

```python
##seaborn plot
x_1 = pd.DataFrame({'var':var,'var_1':var_1})
sns.lineplot(x='var',y='var_1',data=x_1)

plt.show()
```

```python
y_1=sns.load_dataset("penguins")
y_1
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=y_1)

plt.show()
```

```python
sns.lineplot(x='bill_depth_mm',y='body_mass_g',data=y_1,hue='sex',size=30,style='sex',palette='turbo_r',markers=['o','>'],dashes=True)
plt.grid()
plt.title('python')
plt.show()
```

# HISTOGRAM PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("penguins")
var
```

```python
sns.displot(var['bill_length_mm'],bins=[0,10,20,30,40,50,60])  ## displot which will give us the histogram plot
plt.show()
```

```python
sns.displot(var['bill_length_mm'],kde=True,rug=True,color='blue',log_scale=True)
plt.show() ###KDE=Kernal Dencity Estimeter
```

# SCATTER PLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=sns.load_dataset("penguins")
var
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=var)
plt.show()
```

```python
##m={'Male':'*','Female':'o'}
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=var,hue='sex',style='sex',size='sex',sizes=(50,10),palette='turbo_r',alpha=1)
plt.show()
```

# VIOLIN PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.violinplot(x='day',y='total_bill',data=var,hue='time',style='time')
plt.grid()
plt.show() 
```

```python
sns.violinplot(x='day',y='total_bill',data=var,linewidth=2,palette="cool")
plt.show()
```

```python
sns.violinplot(x='time',y='total_bill',data=var,order=['Dinner','Lunch'],inner='quart')
plt.show()
```

```python
##saturation
sns.violinplot(x='day',y='total_bill',data=var,saturation=0.4,color='green')
plt.show()
```

```python
##differentiate by male and female
sns.violinplot(x='day',y='total_bill',data=var,hue='sex',split=True,scale='width')
plt.show()
```

# PAIR PLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.pairplot(var,vars=['total_bill','tip'],hue='sex',hue_order=['Female','Male'])
plt.show()
```

```python
sns.pairplot(var,hue='sex',hue_order=['Female','Male'],x_vars=['total_bill','tip'])
plt.show()
```

```python
sns.pairplot(var,hue='sex',hue_order=['Female','Male'],kind='kde',diag_kind='hist')  ##'scatter','kde','reg'
plt.show()
```

```python
sns.pairplot(var,hue='sex',hue_order=['Female','Male'],markers=['<','*'])
plt.show()
```
